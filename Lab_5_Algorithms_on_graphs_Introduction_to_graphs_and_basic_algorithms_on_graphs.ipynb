```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from google.colab import files
```

Adjacency Matrix Representation

```python
np.random.seed(2022)

nodes = 100
edges = 200

adj_matrix = np.zeros((nodes, nodes))
n_edge = 0

while n_edge < edges:
    i = np.random.randint(0, nodes)
    j = np.random.randint(0, nodes)
    if adj_matrix[i, j] == 0 and i != j:
        adj_matrix[i, j] = 1
        adj_matrix[j, i] = 1
        n_edge += 1

print(adj_matrix)

graph = nx.from_numpy_matrix(adj_matrix)

plt.figure(figsize = (10, 10))
plt.title("Simple Undirected Unweighted Graph")
nx.draw(graph, with_labels = True, node_color = "tab:blue")

plt.savefig("graph_viz.png", dpi = 200, bbox_inches = "tight")
files.download("graph_viz.png")
```

Adjacency List Representation

```python
adj_list = {}

for i in range(nodes):
    adj_list[i] = []
    for j in range(nodes):
        if adj_matrix[i, j] == 1:
            adj_list[i] += [j]

for k in range(10):
    print("{}: {}".format(list(adj_list.keys())[k], adj_list[k]))
```

DFS and BFS Functions

```python
def depth_first(graph, source):
    return nx.dfs_preorder_nodes(
        G = graph,
        source = source
    )

def breadth_first(graph, source, target):
    return nx.bidirectional_shortest_path(
        G = graph,
        source = source,
        target = target
    )
```

Depth-First Search (DFS) Algorithm Implementation

```python
depth_first_args = list(depth_first(graph, 0))

plt.figure(figsize = (10, 10))
plt.title("DFS Result")
nx.draw_shell(graph.subgraph(depth_first_args), with_labels = True, node_color = "tab:blue")

plt.savefig("dfs.png", dpi = 200, bbox_inches = "tight")
files.download("dfs.png")
```

Breadth-First Search (BFS) Algorithm Implementation

```python
breadth_first_args = list(breadth_first(graph, 0, round(len(graph) / 4)))

plt.figure(figsize = (10, 10))
plt.title("BFS Result")
nx.draw_shell(graph.subgraph(breadth_first_args), with_labels = True, node_color = "tab:blue")

print(breadth_first_args)

plt.savefig("bfs.png", dpi = 200, bbox_inches = "tight")
files.download("bfs.png")
```
